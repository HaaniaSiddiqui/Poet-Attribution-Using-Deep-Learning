```
import pandas as pd 
import csv
import requests

# url='/content/drive/MyDrive/data3.csv'
df=  pd.read_csv('data3.csv', encoding='utf-16', header=None, sep=',', names=['text','labels'])
# df=  pd.read_csv(url, encoding='utf-16', header=None, sep=',', names=['text','labels'])
print(df.head())
```

```
from pandas.io.formats.format import DataFrameRenderer
print(set(df["labels"].values))
print(len(set(df["labels"].values)))
print(sum(df["labels"].value_counts()))
```

```
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
labeler = preprocessing.LabelEncoder()
df['labels'] = labeler.fit_transform(df["labels"])
print(max(df['labels']))
df_train, df_test = train_test_split(df, test_size=0.2)
```

```
print(df)
```

# mBERT standard

```
# set use_cuda=False on CPU-only platforms
model = ClassificationModel('bert', 'bert-base-multilingual-uncased', num_labels=15, use_cuda=True, args={
    'reprocess_input_data': True,
    'num_train_epochs': 10,
    'logging_steps': 10,
    'wandb_project': "poet-bert-multilingual",
})
model.train_model(df_train)
```

```
result, model_outputs, wrong_predictions = model.eval_model(df_test)
bads = {}
for pred in wrong_predictions:
    if pred.label in bads:
        bads[pred.label] += 1
    else:
        bads[pred.label] = 1
print("wrong predictions (by input I think?):")
print(bads)
```
